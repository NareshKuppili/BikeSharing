```python
# Building a linear model

y_train = df_train.pop('cnt') # dependent variable is count
X_train = df_train # independent variables

# Importing RFE and LinearRegression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variables limited to 15

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)  # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
# List the columns, rfe support, rfe ranking
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
# List of columns which are supported by rfe assesment

col = X_train.columns[rfe.support_]
col
```

```python
# get variables not supported by rfe
X_train.columns[~rfe.support_]
```

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Build with stats model

# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a df that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
import statsmodels.api as sm

# model 1

# Add a constant
X_train_lm1 = sm.add_constant(X_train_rfe)

# Create a first fitted model
lr1 = sm.OLS(y_train, X_train_lm1).fit()
```

```python
# Parameters dervied from fitted model 1
lr1.params
```

```python
# Checck the summary of model 1
print(lr1.summary())
```

## Dropping the variable and updating the model:

The p-value for each term tests the null hypothesis that the coefficient is equal to zero (no effect). A low p-value (< 0.05) indicates that you can reject the null hypothesis. Predictor that has a low p-value is likely to be a meaningful addition to th model. So we can drop variables having p value > 0.05

```python
# drop the column 'atemp' from the train set
X_train_new = X_train_rfe.drop(["atemp"], axis = 1)

# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_new.columns
vif['VIF'] = [variance_inflation_factor(X_train_new.values, i) for i in range(X_train_new.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
